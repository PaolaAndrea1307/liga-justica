### **Ejercicio**

Crear dos vectores de diez numeros aleatorios enteros [np.random.randint](https://docs.scipy.org/doc/numpy-1.15.0/reference/generated/numpy.random.randint.html) entre 1 y 10 y realizar las siguientes operaciones:
- sumarlos
- concatenar los dos arreglos
- encontrar la sumatoria de los valores
- encontrar la media de los valores
- imprimir los numeros mayores que 5
- determinar las posiciones de los numeros que son iguales en los arreglos



````python

# Crear vectores
vec1 = np.random.randint(1, 11, 10)
vec2 = np.random.randint(1, 11, 10)
print(vec1)
print(vec2)

# Sumar vetores
suma = vec1 + vec2
print(suma)

# Concatenar vectores
concatVec = np.concatenate((vec1, vec2))
print(concatVec)

# Sumatoria de valores
sumatoria = np.sum(concatVec)
print(sumatoria)

# Media de los valroes
media = np.mean(concatVec)
print(media)

# 5. Imprimir los números mayores que 5
mayores = concatVec[concatVec > 5]
print(mayores)

# Las posiciones donde los valores de ambos vectores son iguales
iguales = np.where(vec1 == vec2)
print(iguales)
````

````python
import pandas as pd
from IPython.display import display, HTML
data = pd.read_csv('guia_fasecolda.csv')

print(data.columns)

automoviles_automaticos = data[(data['Clase'] == 'AUTOMOVIL') & (data['TipoCaja'] == 'AT') & (data['Marca'].isin(["TOYOTA", "RENAULT", "CHERVROLET", "KIA", "MAZDA"]))]

descripcion_estadistica = automoviles_automaticos.describe()
print("\nDescripción Estadística de los Automoviles de caja automatica:")
print(descripcion_estadistica)

mode_marca = automoviles_automaticos['Marca'].mode()
print("\nModa de la marca de los carros automaticos")
print(mode_marca)

median_2018 = automoviles_automaticos['2018'].median()
print("\nMediana de el precio de los carros automaticos en 2018")
print(median_2018)

display(HTML((automoviles_automaticos.head().to_html())))
````

````python

%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

plt.figure()
automoviles_automaticos[['2018']].plot(kind = 'box',ylim=(0,200000))
plt.title('Precios de 2018')
plt.xlabel("Año")
plt.ylabel("Precio")
plt.show()

````

````python
%matplotlib inline

df_2018 = automoviles_automaticos[['Marca', 'Potencia', '2018']]

df_2018 = df_2018.dropna(subset=['Potencia', '2018'])

plt.figure(figsize=(10, 6))

colors = plt.get_cmap('Set1', len(df_2018['Marca'].unique()))

for i, marca in enumerate(df_2018['Marca'].unique()):
    marca_data = df_2018[df_2018['Marca'] == marca]
    plt.scatter(marca_data['Potencia'], marca_data['2018'], color=colors(i / len(df_2018['Marca'].unique())), label=marca, alpha=0.6)

plt.title("Precio vs Potencia por Marca (2018)")
plt.xlabel("Potencia")
plt.ylabel("Precio en 2018")

plt.xlim(0, df_2018['Potencia'].max() + 10)
plt.ylim(0, df_2018['2018'].max() + 10000)

plt.legend(title='Marca', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()

plt.show()
````

````python
filtro_chevrolet = data[(data["Marca"] == "CHEVROLET") & (data["Clase"] == "AUTOMOVIL")]
filtro_chevrolet
conteo_marcas = filtro_chevrolet['Referencia1'].value_counts()
top_5_marcas = conteo_marcas.head(5)

plt.figure(figsize=(10, 6))
top_5_marcas.plot(kind='bar')
plt.title('Top 5 de Marcas con Mayor Cantidad de Registros')
plt.xlabel('Marca')
plt.ylabel('Cantidad de Registros')
plt.show()
````
